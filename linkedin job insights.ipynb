```python
# Plotting a histogram of seniority_level
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='seniority_level')
plt.title('Seniority Level Distribution')
plt.xlabel('Seniority Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting a bar chart of hiring_status
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='hiring_status')
plt.title('Hiring Status Distribution')
plt.xlabel('Hiring Status')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting a bar chart of employment_type
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='employment_type')
plt.title('Employment Type Distribution')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.show()
```
